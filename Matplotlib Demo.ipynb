### Scatter plot

```python
x=[1, 1.5, 2, 2.5, 3, 3.5, 3.6]
y=[7.5, 8, 8.5,9,9.5,10,10.5]
x2=[8, 8.5,9,9.5,10,10.5,11]
y2=[3, 3.5,3.7,4,4.5,5,5.2]

plt.scatter(x,y, label='high income group', color='r')
plt.scatter(x2,y2, label='low income group', color='g')
plt.title('Scatter Plot')
plt.xlabel('savings*100')
plt.ylabel('income*1000')
plt.grid(True)
plt.legend()
plt.show()
```

### Bar Graph

```python
subject=['Maths','Science','Computer','Economics']
average_marks=[78, 89, 68, 93]

plt.bar(subject, average_marks, color='green')
plt.title('Bar Graph')
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.show()
```

```python
subject=['Maths','Science','Computer','Economics']
boys_average_marks=[78, 89, 68, 93]
girls_average_marks=[75, 92, 76, 98]

index=np.arange(4)
width=0.30

plt.bar(index, boys_average_marks, width, color='blue', label='Boys Average Marks')
plt.bar(index+width, girls_average_marks, width, color='red', label='Girls Average Marks')
plt.title('Bar Graph Stacked Horizontally')
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.legend()
plt.show()
```

```python
subject=['Maths','Science','Computer','Economics']
boys_average_marks=[78, 89, 68, 93]
girls_average_marks=[75, 92, 76, 98]

index=np.arange(4)
width=0.30

plt.bar(index, boys_average_marks, width, color='blue', label='Boys Average Marks')
plt.bar(index, girls_average_marks, width, color='red', label='Girls Average Marks', bottom=boys_average_marks)
plt.title('Bar Graph Stacked Horizontally')
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.legend()
plt.xticks(index, subject)
plt.show()
```

## Pie Charts

```python
x=np.random.randint(0, 100, 5)

plt.pie(x)
```

```python
x
```

```python
list_val=['Saving','Income','Expend','Misc','Others']
plt.pie(x, labels=list_val, startangle=90, shadow=True)
```

```python
list_val=['Saving','Income','Expend','Misc','Others']
plt.pie(x, labels=list_val, startangle=90, shadow=True, autopct="%1.1f")
```

```python
list_val=['Saving','Income','Expend','Misc','Others']
plt.pie(x, labels=list_val, startangle=90, shadow=True, autopct="%1.1f", explode=(0,0.1,0,0,0.1))
```

### 3D Plots

```python
from mpl_toolkits import mplot3d

height=np.array([167,170,145,165,155,180,166,147,159,167,172,181,190])
width=np.array([86,74,66,78,79,90,72,70,83,66,67,85,76])

plt.xlim(140, 200)
plt.ylim(60,100)

ax=plt.axes(projection='3d')
ax.scatter3D(height, width)
ax.set_xlabel('Height')
ax.set_ylabel('Width')
plt.show()
```

### Multiple Plots

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.pie(np.random.randint(0,100,5), autopct="%1.1f", explode=(0,0,0,0,0.1))
ax2.scatter(np.random.randint(0,50,10), np.random.randint(100,200,10))
ax2.grid(True)
ax3.plot([n for n in range(50,200,5)], color='r')
ax4.bar(subject, boys_average_marks, label='average marks', color='g')
ax4.legend()
```

### Subplot2Grid

```python
gridsize=(3,2)
fig=plt.figure(figsize=(12,8))
ax1=plt.subplot2grid(gridsize,(0,0), colspan=2, rowspan=2)
ax2=plt.subplot2grid(gridsize,(2,0))
ax3=plt.subplot2grid(gridsize,(2,1))

ax1.pie(np.random.randint(0,100,5), autopct="%1.1f", explode=(0,0,0,0,0.1))
ax2.scatter(np.random.randint(0,50,10), np.random.randint(100,200,10))
ax2.grid(True)
ax3.plot([n for n in range(50,200,5)], color='r')
```
