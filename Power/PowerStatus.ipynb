```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
%matplotlib qt
```

```python
with open('PowerData.csv') as f:
    pdata = pd.DataFrame(json.loads(line) for line in f)
```

```python
pdata
```

```python
pdata_ev = pdata.loc[pdata['Type'] == 'EVENT']
```

```python
pdata_ev
```

```python
pdata_ev['Timestamp'] = pd.to_datetime(pdata_ev['Timestamp'])
```

```python
pdata_ev_days = pdata_ev.set_index('Timestamp')
```

```python
pdata_ev_days_grp = pdata_ev_days.groupby(pd.TimeGrouper(freq='D'))
```

```python
type(pdata_ev_days_grp)
```

```python
pdata_ev_days_grp.describe()
```

```python
pdata_ev_days_grp_count = pdata_ev_days_grp.count()
```

```python
pdata_ev_days_grp_count
```

```python
# Plot Number of Events to Date
```

```python
plt.plot(pdata_ev_days_grp_count.index, pdata_ev_days_grp_count.Type,marker='o')
plt.xlabel('Date')
plt.ylabel('Number of Events')
```

```python
#Data on 2016-08-28
pdata_ev_days_grp_28 = pdata_ev_days_grp.get_group('2016-08-28')
```

```python
pdata_ev_days_grp_28
```

```python
pdata_ev_days_grp_28['Time'] = pdata_ev_days_grp_28.index
```

```python
pdata_ev_days_grp_28['delta'] = (pdata_ev_days_grp_28['Time']-pdata_ev_days_grp_28['Time'].shift()).fillna(0)
```

```python
pdata_ev_days_grp_28
```

```python
pdata_ev_days_grp_28 = pdata_ev_days_grp_28.replace('POWER_CONNECTED', 1)
pdata_ev_days_grp_28 = pdata_ev_days_grp_28.replace('POWER_DISCONNECTED', 0)
```

```python
pdata_ev_days_grp_28
```

```python
#Plot Event vs Time
plt.plot(pdata_ev_days_grp_28.Time, pdata_ev_days_grp_28.Status,marker='o')
plt.xlabel('Time Duration')
plt.ylabel('Status')
```

```python
pdata_ev_days_grp_28_powercutduration = pdata_ev_days_grp_28.loc[pdata_ev_days_grp_28['Status'] == 1]
```

```python
pdata_ev_days_grp_28_powercutduration
```

```python
#Maximum duration of Power cut
max(pdata_ev_days_grp_28_powercutduration.delta)
```
